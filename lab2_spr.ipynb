```python
import numpy as np
import matplotlib.pyplot as plt

# Common parameters
fs = 1000  # Sampling frequency in Hz
T = 1  # Duration in seconds
t = np.linspace(0, T, int(fs * T), endpoint=False)  # Time vector
N = len(t)  # Number of samples
frequencies = np.linspace(-fs/2, fs/2, 1000)  # Frequency vector for DTFT
freq_dft = np.fft.fftshift(np.fft.fftfreq(N, 1/fs))  # Frequency vector for DFT

# Question 1: Sinusoidal Signal
# a) Generate sinusoidal signal
f = 5  # Frequency in Hz
x_sine = np.sin(2 * np.pi * f * t)

# b) Plot time-domain waveform
plt.figure(figsize=(12, 12))
plt.subplot(3, 1, 1)
plt.plot(t, x_sine)
plt.title('Q1: Time-Domain Sinusoidal Signal (5 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# c) Compute and plot DTFT
X_dtft_sine = np.zeros(len(frequencies), dtype=complex)
for i, freq in enumerate(frequencies):
    omega = 2 * np.pi * freq / fs
    X_dtft_sine[i] = np.sum(x_sine * np.exp(-1j * omega * np.arange(N)))
plt.subplot(3, 1, 2)
plt.plot(frequencies, np.abs(X_dtft_sine))
plt.title('Q1: DTFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

# d) Compute and plot DFT
X_dft_sine = np.fft.fftshift(np.fft.fft(x_sine))
plt.subplot(3, 1, 3)
plt.stem(freq_dft, np.abs(X_dft_sine))
plt.title('Q1: DFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.tight_layout()
plt.show()

# Question 2: Composite Signal
# a) Generate composite signal
x_comp = np.sin(2 * np.pi * 5 * t) + 0.5 * np.sin(2 * np.pi * 10 * t)

# b) Plot time-domain waveform
plt.figure(figsize=(12, 12))
plt.subplot(3, 1, 1)
plt.plot(t, x_comp)
plt.title('Q2: Time-Domain Composite Signal (5 Hz + 10 Hz)')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# c) Compute and plot DTFT
X_dtft_comp = np.zeros(len(frequencies), dtype=complex)
for i, freq in enumerate(frequencies):
    omega = 2 * np.pi * freq / fs
    X_dtft_comp[i] = np.sum(x_comp * np.exp(-1j * omega * np.arange(N)))
plt.subplot(3, 1, 2)
plt.plot(frequencies, np.abs(X_dtft_comp))
plt.title('Q2: DTFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

# d) Compute and plot DFT
X_dft_comp = np.fft.fftshift(np.fft.fft(x_comp))
plt.subplot(3, 1, 3)
plt.stem(freq_dft, np.abs(X_dft_comp))
plt.title('Q2: DFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.tight_layout()
plt.show()

# Question 3: Exponentially Decaying Signal
# a) Generate exponentially decaying signal
alpha = 2  # Decay constant
x_exp = np.exp(-alpha * t) * np.sin(2 * np.pi * 5 * t)

# b) Plot time-domain waveform
plt.figure(figsize=(12, 12))
plt.subplot(3, 1, 1)
plt.plot(t, x_exp)
plt.title('Q3: Time-Domain Exponentially Decaying Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# c) Compute and plot DTFT
X_dtft_exp = np.zeros(len(frequencies), dtype=complex)
for i, freq in enumerate(frequencies):
    omega = 2 * np.pi * freq / fs
    X_dtft_exp[i] = np.sum(x_exp * np.exp(-1j * omega * np.arange(N)))
plt.subplot(3, 1, 2)
plt.plot(frequencies, np.abs(X_dtft_exp))
plt.title('Q3: DTFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

# d) Compute and plot DFT
X_dft_exp = np.fft.fftshift(np.fft.fft(x_exp))
plt.subplot(3, 1, 3)
plt.stem(freq_dft, np.abs(X_dft_exp))
plt.title('Q3: DFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.tight_layout()
plt.show()

# e) Analysis
print("Q3e Analysis: The time-domain signal decays exponentially, reducing amplitude over time. "
      "In the frequency domain, the DTFT and DFT show a broadened peak around 5 Hz due to the "
      "exponential decay, which introduces a range of frequencies (spectral spreading). The DFT "
      "discretizes this spectrum, showing similar broadening but at discrete frequency bins.")

# Question 4: Rectangular Pulse Signal
# a) Generate rectangular pulse signal
pulse_duration = 0.2  # Pulse duration in seconds
x_rect = np.zeros_like(t)
x_rect[t < pulse_duration] = 1  # Pulse is 1 for t < 0.2s, else 0

# b) Plot time-domain waveform
plt.figure(figsize=(12, 12))
plt.subplot(3, 1, 1)
plt.plot(t, x_rect)
plt.title('Q4: Time-Domain Rectangular Pulse Signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.grid(True)

# c) Compute and plot DTFT
X_dtft_rect = np.zeros(len(frequencies), dtype=complex)
for i, freq in enumerate(frequencies):
    omega = 2 * np.pi * freq / fs
    X_dtft_rect[i] = np.sum(x_rect * np.exp(-1j * omega * np.arange(N)))
plt.subplot(3, 1, 2)
plt.plot(frequencies, np.abs(X_dtft_rect))
plt.title('Q4: DTFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

# d) Compute and plot DFT
X_dft_rect = np.fft.fftshift(np.fft.fft(x_rect))
plt.subplot(3, 1, 3)
plt.stem(freq_dft, np.abs(X_dft_rect))
plt.title('Q4: DFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)
plt.tight_layout()
plt.show()

# e) Analysis
print("Q4e Analysis: The time-domain rectangular pulse is finite and abrupt, leading to a sinc-shaped "
      "frequency spectrum in the DTFT, with a main lobe width inversely proportional to the pulse duration (0.2s). "
      "The DFT discretizes this sinc function, showing the same characteristic lobes but sampled at discrete frequencies. "
      "The abrupt transitions in the time domain cause high-frequency components in the spectrum.")
```
